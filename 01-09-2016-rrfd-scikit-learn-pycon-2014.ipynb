```python
# Now lets take some time to see what the data looks like

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(style='darkgrid')

def plot_iris_projection(x_index, y_index):
    # this formatter will label the colorbar with the correct target names
    formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

    # c --> color or sequence of color, optional, default.
    sns.plt.scatter(iris.data[:, x_index], iris.data[:, y_index],
                c=iris.target)
    plt.colorbar(ticks=[0, 1, 2], format=formatter)
    plt.xlabel(iris.feature_names[x_index])
    plt.ylabel(iris.feature_names[y_index])
    
plot_iris_projection(0, 3)
```

```python
# Lets import some other datasets and see what we can learn

from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# This is to show the shape
print digits.data.shape

# This is to save the column and rows
# BTW there are a lot of feature names
n_samples, n_features = digits.data.shape
```

```python
digits.keys()
```

```python
# But take a look at this.
# Looks like we have two different versions of the data. 8*8=64.
print digits.images.shape
```

```python
# Lets reshape the data.
digits_test = digits.images.reshape((1797, 64))
digits_test
```

```python
# Let's test the data
# Looks like it evaluates the same.
print np.all(digits_test == digits.data)
```

```python
# Great lets visualize the data so we can see what is going on.

fig = plt.figure(figsize=(6,6)) # Remember this is in inches
fig.subplots_adjust(left=0, right = 1, bottom = 0, top = 1)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # Label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```
