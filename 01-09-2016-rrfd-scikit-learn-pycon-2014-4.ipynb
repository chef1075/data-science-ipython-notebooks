We're looking at unsupervised learning at the moment. So it is PCA and K-Means.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

PCA is a powerful UL method for dimensionality reduction in data. It's easiest to visualize by looking at a two-dimenional dataset.

```python
np.random.seed(1)
X = np.dot(np.random.random(size=(2, 2)), np.random.random(size=(2, 200))).T
plt.plot(X[:, 0], X[:, 1], 'og')
plt.axis('equal')
```

There is an trend to the data...but we should look at it a bit closer.

What PCA is going to do here, is look at the Principal Axes in the data, and explain - to us - how important those axes are in decribing the data distribution.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_)
print(pca.components_)

# Will give a array as an answer so we need to visualize this
```

```python
# This is not the X from the first graph, but the one from the PCA
plt.plot(X[:, 0], X[:, 1], 'og', alpha = 0.3)
plt.axis('equal')

# zip() --> This function returns a list of tuples, where the i-th tuple 
# contains the i-th element from each of the argument 
# sequences or iterables
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    plt.plot([0, v[0]], [0, v[1]], '-k', lw=3)
```

The direction is very important, while the other direction is not. this show us that the second principal component could be ignored ... without much loss of information! Let's take a look if we only keep 95% var.

```python
# Re do PCA with 95% VAR
clf = PCA(.95)
X_trans = clf.fit_transform(X)
print(X.shape)
print(X_trans.shape)
```

Compression has occured and it looks like it is right on track to be correct.

```python
X_new = clf.inverse_transform(X_trans)

# The orginal graph
plt.plot(X[:, 0], X[:, 1], 'og', alpha=0.2)
plt.plot(X_new[:, 0], X_new[:, 1], 'og', alpha=0.8)
plt.axis('equal');
```

We removed a lot of variance from the data, and it still keeps its shape. The fit is ALOT better.

```python
# Lets take a look at high-dimensional data (a lot of axis) but now lets
# add some PCA to it.

from sklearn.datasets import load_digits
digits = load_digits()

X = digits.data
y = digits.target
```

```python
pca = PCA(2) # project from 64 to 2 dimensions
Xproj = pca.fit_transform(X)

# Looking at how the data is structured
print(X.shape)
print(Xproj.shape)
```

```python
plt.scatter(Xproj[:, 0], Xproj[:, 1], c=y)
plt.colorbar();
```

This gives us an idea of the relationship between the digits. Essentially, we have found the optimal reotation in 64-dimensional space that allows for us to see the layout of the digits without reference to the labels.

But how much info have we thrown away? Can we figure this out by looking at the variances?!

```python
pca = PCA(64).fit(X)
plt.semilogx(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of compnents')
plt.ylabel('cumulative explained variance')
```

What does this mean? we need about 20 components to retain 90% of the varaiance. Looking at this plot for a high-dimensional dataset can help you understand the level of redudency present in pultipule oberservations.

Ohhh exciting ... now on to K Mean clustering! Another type of UL that is finding clusters in data based on the data attributes alones (not the labels).

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples = 900, centers = 5,
                  random_state = 1, cluster_std = 0.9)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

You can kind of see clusters, but they is because you plotted it. What if you just had raw data?

Let's take a look at Expectation Max (EM) procedure which scikit-learn omplements, so that KMeans can be solved quickly.

```python
from sklearn.cluster import KMeans
est = KMeans(5) # Becausse I have 5 clusters
est.fit(X)
y_kmeans = est.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50);
```

```python
from sklearn.cluster import KMeans
est = KMeans(8) # Lets look for more clusters?
est.fit(X)
y_kmeans = est.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50);
```

```python
# n_clusters --> The number of clusters to form as well as the number of centroids to generate.
est = KMeans(n_clusters=10)

# fit_predict --> Compute cluster centers and predict cluster index for each sample.
clusters = est.fit_predict(digits.data)
est.cluster_centers_.shape

# This shows what the algo thinks without visualizing anything.
```

```python
fig = plt.figure(figsize=(8, 3))
for i in range(10):
    ax = fig.add_subplot(2, 5, 1 + i, xticks=[], yticks=[])
    #cmap --> colour map
    ax.imshow(est.cluster_centers_[i].reshape((8,8)), cmap=plt.cm.binary)
```

Look .. even without the laels KMeans is able to find clusters whose means are recognizable digits .... except for 8.

For good measure let's use our PCA visualization and look at the true cluster labels and K-mean cluster labels.

```python
X =PCA(2).fit_transform(digits.data)

# there are going to be two graphs
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].scatter(X[:,0], X[:,1], c=clusters)
ax[1].scatter(X[:,0], X[:,1], c=digits.target);
```

What does this mean? Well it looks like the KMeans closely mirrors the true clustering of the data.

Ok, let take a look at KMeans ... but this time for colour compression. Because an image could have millions of colors, but at what point do you make a comparison between certain colours?

```python
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg")
plt.imshow(china);
```

Fancy!

```python
china.shape
# print type(china)
```

The image is 3D because of height, width, and RGB

```python
# Lets reshape this so it looks like a typical scikit-learn input

# reshape() --> create a new array shape
X = (china / 255.0).reshape(-1, 3)
print(X.shape)
```

This means we have 273280 in three demintions! Now lets use KMean to compress the 256^3 colors into a smaller number (say 64 colors). Basically, you want to find color clusters in the data, and create a new image where the true input color of the closest cluster.

```python
from sklearn.cluster import KMeans


def compress_image(image, n_colors):
    """Compress an image

    Parameters
    ==========
    image : numpy array
        array of shape (height, width, 3) with values between 0 and 1
    n_colors : integer
        the number of colors in the final compressed image
        (i.e. the number of KMeans clusters to fit).
        
    Returns
    =======
    new_image : numpy array
        array representing the new image, compressed via KMeans clustering.
        It has the same shape as the input image, but contains only
        ``n_colors`` distinct colors.
    """
    
    X = image.reshape(-1, 3) / 255.
    model = KMeans(n_colors)
    labels = model.fit_predict(X)
    colors = model.cluster_centers_
    new_image = colors[labels].reshape(image.shape)
    return (255 * new_image).astype(np.uint8)

# create and plot the new image
new_image = compress_image(china, 64)
plt.imshow(new_image);
```

You can see how the image has been compressed from 256 to 64, and the machine learning algo simply bunched together the KMean clusters to make sure the visual quality of the image was not completely distroyed.

I.e. you still know what it is.
