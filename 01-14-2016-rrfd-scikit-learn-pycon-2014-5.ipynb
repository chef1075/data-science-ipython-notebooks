```python
# Lets use a SVM and see how well the model can classify this data.

from sklearn.svm import SVC

# Does not necicarily need to be linear ... 
# Make sure to remember to set the classifier as the SVC,
# as this is simply an application of the function to a varaiable "clf"
clf = SVC()

# What you want to do is split the data (X and y) into a training and test set
# train_test_split() --> splits the data, duh
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)

# Now you fit the model to the training
clf.fit(Xtrain, ytrain)

# Compute y_pred, the predicted labels of the test data
ypred = clf.predict(Xtest)
```

```python
# Now lets take a look a the precision, the recall, and f1 score.

from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

# This is one line of code, rather than 2 in the section above
clf = SVC().fit(X_train, y_train)

# Notice how you're putting all the data into the CLF but with
# different transformations and data segmentations.
y_pred = clf.predict(X_test)

# Now we can look directly at the scores
print "The Accuracy of the SVC is:", metrics.accuracy_score(y_test, y_pred)
print "The Precision of the SVC is:", metrics.precision_score(y_test, y_pred)
print "The Recall of the SVC is:", metrics.recall_score(y_test, y_pred)
print "The F1 score is:", metrics.f1_score(y_test, y_pred);
```

What do these mean?

Accuracy --> correct label/total samples

Precision --> true positives/(true positives + false positives)

Recall --> true positives/(true positives + flase negatives)

F1 = 2[(precision*recall)/(precision+recall)]

```python
# We could go through the above ... or we could look at the 
# data with a quick line of code

print metrics.classification_report(y_test, y_pred,
                                    target_names=['background', 'foreground'])
```

Now we've gone over the basics of the metrics, validation, and cross validation. Let's go further in depth regarding model seleciton.

Let's take a look at : if our estimator is under performing, how should we move forward.


```python
# The array at the end, makes X_test a column vector.
X_test = np.linspace(-0.1, 1.1, 500)[:, None]
```

```python
def test_func(x, err=0.5):
    y = 10 - 1. / (x + 0.1)
    if err > 0:
        y = np.random.normal(y, err)
    return y

def make_data(N=40, error=1.0, random_seed=1):
    # randomly sample the data
    np.random.seed(1)
    X = np.random.random(N)[:, np.newaxis]
    y = test_func(X.ravel(), error)
    
    return X, y

X, y = make_data(40, error=1)
plt.scatter(X.ravel(), y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# same process as the classifier from SVMs
model.fit(X, y)

# The data has been split
y_test = model.predict(X_test)

# revel() --> Return a contiguous flattened array.
plt.scatter(X.ravel(), y)
plt.plot(X_test.ravel(), y_test)
print "mean squared error:", metrics.mean_squared_error(model.predict(X), y)
```

This line KINDA fits the model, but I would say it under-fits the data. Lets try to improve this by creating more complicated models ... and we can do this by adding DoF and computing a polynomial regression over the inputs.

```python
class PolynomialRegression(LinearRegression):
    """Simple Polynomial Regression to 1D data"""
    def __init__(self, degree=1, **kwargs):
        self.degree = degree
        LinearRegression.__init__(self, **kwargs)
        
    def fit(self, X, y):
        if X.shape[1] != 1:
            raise ValueError("Only 1D data valid here")
        Xp = X ** (1 + np.arange(self.degree))
        return LinearRegression.fit(self, Xp, y)
        
    def predict(self, X):
        Xp = X ** (1 + np.arange(self.degree))
        return LinearRegression.predict(self, Xp)
```

Now lets look at a quadratic curve with the data.

```python
model = PolynomialRegression(degree=2)
model.fit(X, y)
y_test = model.predict(X_test)

plt.scatter(X.ravel(), y)
plt.plot(X_test.ravel(), y_test)
print "mean squared error:", metrics.mean_squared_error(model.predict(X), y)
```

The MSE reduces, so there is a better fit!

Now lets increase the polynomial values ...

```python
model = PolynomialRegression(degree=15)
model.fit(X, y)
y_test = model.predict(X_test)

plt.scatter(X.ravel(), y)
plt.plot(X_test.ravel(), y_test)
plt.ylim(-4, 14)
print "mean squared error:", metrics.mean_squared_error(model.predict(X), y)
```

Hummm ... this is now overfitted, which is no longer helpful.

Now lets detect some over fitting

```python
degrees = np.arange(1, 30)

X, y = make_data(100, error=1.0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

training_error = []
test_error = []
mse = metrics.mean_squared_error

for d in degrees:
    model = PolynomialRegression(d).fit(X_train, y_train)
    training_error.append(mse(model.predict(X_train), y_train))
    test_error.append(mse(model.predict(X_test), y_test))
    
# note that the test error can also be computed via cross-validation
# going to produce a line graph
plt.plot(degrees, training_error, label='training')
plt.plot(degrees, test_error, label='test')
plt.legend()
plt.title('Typical Bias/Variance Plot')
plt.xlabel('degree')
plt.ylabel('MSE');
```

Optimal model here is when the test error is minimized

Now lets just look at learning curves.

We need to see the EXACT turning point of the tradeoff between bias and veriance is highly dependent on the number of training points used. Here we'll illustrate the use of learning curves, which dispaly this property.

The idea is to plot the MSE for the training and test set as a function of the number of training points.

```python
# Creating random data again.
X, y = make_data(200, error=1.0)
degree = 3

# Same way to split data as before.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

N_range = np.linspace(15, X_train.shape[0], 20).astype(int)

def plot_learning_curve(degree=3):
    training_error = []
    test_error = []
    mse = metrics.mean_squared_error
    
    for N in N_range:
        XN = X_train[:N]
        yN = y_train[:N]
    
        model = PolynomialRegression(degree).fit(XN, yN)
        training_error.append(mse(model.predict(XN), yN))
        test_error.append(mse(model.predict(X_test), y_test))
      
    plt.plot(N_range, training_error, label='training')
    plt.plot(N_range, test_error, label='test')
    plt.plot(N_range, np.ones_like(N_range), ':k')
    plt.legend()
    plt.title('degree = {0}'.format(degree))
    plt.xlabel('num. training points')
    plt.ylabel('MSE')
    
plot_learning_curve(3)
```

What does this mean? For very few training points, there is a LARGE seperation between the training and test error, which indicates OVER FITTING. Given the same model, for a large number of training points, the training and testing errors converge, which indicates *maybe* under-fitting.

If you want to reduce the MSE to a nominal value of 1.0 then adding more sample will never get you there. For degree (d = 3) we've already converaged. 

What about d1? or 2?

```python
plot_learning_curve(1)

# Nope
```

```python
plot_learning_curve(2)

# Nope
```

```python
plot_learning_curve(5)

# Nope
```

We did it! After degrees of 5 we have converged.

Thus we can bring together the curves by adding more points, but you can bring down the convergence level down by ***adding complexity*** to the model.

This gives you a way to explore the data: should you add more objects to get a larger training set, or more attributes of each object in order to improve the model?
